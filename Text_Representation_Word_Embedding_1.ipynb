Bag of Words

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({"text":["people watch Aman",
                           "dswithbappy watch Aman",
                           "people write comment",
                           "Aman write comment"],"output":[1,1,0,0]})
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
```

```python
#Vocabulary
print(cv.vocabulary_)
```

```python
bow
```

```python
bow.toarray()
```

```python
print(bow[0].toarray())
print(bow[1].toarray())
print(bow[2].toarray())
```

```python
# new
cv.transform(['Aman watch dswithAman']).toarray()
```
