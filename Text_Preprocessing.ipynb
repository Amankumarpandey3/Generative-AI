Remove HTML tags


```python
import re
def remove_html_tags(text):
  pattern=re.compile('<.*?>')
  return pattern.sub(r'',text)
```

remove_url

```python
def remove_url(text):
  pattern = re.compile(r'https?://\S+|www\.\S+')
  return pattern.sub(r'',text)
```

Punctuation handling

```python
import string,time
string.punctuation
```

```python
exclude = string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  return text
```

```python
text='string. With. Punctuation?'
```

```python
start=time.time()
print(remove_punc(text))
time1=time.time()-start
print(time1*50000)
```

Best way


```python
def remove_punc1(text):
  return text.translate(str.maketrans('','',exclude))
```

```python
start = time.time()
remove_punc1(text)
time2 = time.time() - start
print(time2*50000)
```

chat_conversion handle

```python
chat_words = {
    'AFAIK':'As Far As I Know',
    'AFK':'Away From Keyboard',
    'ASAP':'As Soon As Possible'
}


{
    "FYI": "For Your Information",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "BTW": "By The Way",
    "OMG": "Oh My God",
    "IMO": "In My Opinion",
    "LOL": "Laugh Out Loud",
    "TTYL": "Talk To You Later",
    "GTG": "Got To Go",
    "TTYT": "Talk To You Tomorrow",
    "IDK": "I Don't Know",
    "TMI": "Too Much Information",
    "IMHO": "In My Humble Opinion",
    "ICYMI": "In Case You Missed It",
    "AFAIK": "As Far As I Know",
    "BTW": "By The Way",
    "FAQ": "Frequently Asked Questions",
    "TGIF": "Thank God It's Friday",
    "FYA": "For Your Action",
    "ICYMI": "In Case You Missed It",
}
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
chat_conversion('Do this work ASAP')
```

Stopswords

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
  new_text=[]
  for word in text.split():
    if word in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(word)
  x=new_text[:]
  new_text.clear()
  return "".join(x)
```

Stemmer

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample = 'walk walks walking walked'
stem_words(sample)
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')
wordnet_lemmatizer=WordNetLemmatizer()

sentence='Under the fading lantern light, Mira discovered an old notebook buried beneath dusty books. Each page held sketches of places sheâ€™d never seenâ€”forests, oceans, skies alive with stars. Though she couldnâ€™t travel, she traced the lines with wonder. In that quiet room, imagination became her passport, and every drawing carried her farther than footsteps ever could.'
punctuations='?:!.,;'
sentence_words = nltk.word_tokenize(sentence)
for word in sentence_words:
  if word in punctuations:
    sentence_words.remove(word)

sentence_words
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words:
  print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word,pos='v')))
```
